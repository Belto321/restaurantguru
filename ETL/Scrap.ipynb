## Se abordan los ejemplos para poder realizar web scraping de Google Maps y sus Reviews

```python
import pandas as pd
from clean_functions.scrape_tk import GoogleMapsScraper, GoogleReviewsScraper
```

### Para los restaurantes (negocios)

```python
# Datos de prueba
data = {
    "state_id": ["FL", "NC"],
    "city": ["Cape Coral", "Huntersville"],
    "lat": ["26.6443", "35.4060"],
    "lng": ["-81.9957", "-80.8716"],
    "zips": ["33990 33991", "28078"]
}

df_info_states = pd.DataFrame(data)
api_key = "API-KEY"
gms = GoogleMapsScraper(api_key)
df_results = gms.scrape_google_maps(df_info_states)
df_results.to_parquet("google_maps_data.parquet")
```

```python
df_try = pd.read_parquet("google_maps_data.parquet")
```

```python
df_try2 = df_try.iloc[0:2]
df_try2
```

### Para las reviews de 2 negocios
#### Tener en cuenta que se trae las últimas reviews que se realizan a los locales

```python
api_key = "API-KEY"
grs = GoogleReviewsScraper(api_key)


reviews_df = grs.process_reviews(df_try2)
grs.save_reviews_to_parquet(reviews_df, "reviews.parquet")
```
